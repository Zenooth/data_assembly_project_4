```python
firefox = r'C:\Users\User\Downloads\geckodriver-v0.20.1-win64\geckodriver'
login = 'https://jobscentral.com.sg/login'
url = 'https://jobscentral.com.sg'
search_terms = ['data scientist', 'data analyst', 'business intelligence', 'business analyst', 'data science', 'language processing', 'software engineer']
```

```python
profile = webdriver.FirefoxProfile()
profile.set_preference("permissions.default.image", 2)
```

```python
job_links = []
driver = webdriver.Firefox(executable_path=firefox, firefox_profile=profile)
driver.get(login)
sleep(1)

username = driver.find_element_by_id("inputEmail")
password = driver.find_element_by_id("inputPassword")
username.send_keys("goh.yu.xuan@gmail.com")
password.send_keys("Pa55worD")
driver.find_element_by_id("submitBtn").click()

sleep(5)

# driver.(url)

for search_term in search_terms:
    search = driver.find_element_by_xpath('//div[1]/div/input')
    search.clear()
    search.send_keys(search_term)
    search.send_keys(Keys.RETURN)

    sleep(2)

    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')

    for item in soup.find_all(attrs={"class": "job-title"}):
        if item.get('href') not in job_links:
            job_links.append(item.get('href'))
          
    a = soup.find(attrs={"aria-label": "Next"})
    
    while a:
        urlend = a.get('href')
        driver.get(url+urlend)
        sleep(3)
        html = driver.page_source
        soup = BeautifulSoup(html, 'lxml')
        a = soup.find(attrs={"aria-label": "Next"})
        if urlend == a.get('href'):
            break
        for item in soup.find_all(attrs={"class": "job-title"}):
            if item.get('href') not in job_links:
                job_links.append(item.get('href'))
driver.close()
```
